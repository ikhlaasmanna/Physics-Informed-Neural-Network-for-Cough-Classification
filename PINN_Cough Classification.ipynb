```python
%matplotlib inline
import os
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchaudio
import torchaudio.transforms as T
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
from torch.amp import GradScaler, autocast

# Configuration
config = {
    "data_dir": "/home/chetak/Speech_Mixing/Cough_Augmented_Dataset",
    "sample_rate": 16000,
    "n_mels": 128,
    "fixed_length": 512,
    "batch_size": 32,
    "epochs": 400,
    "learning_rate": 1e-4,
    "warmup_epochs": 5,
    "patience": 20,
    "num_workers": 4,
    "dropout": 0.3,
    "num_layers": 8,
    "embed_dim": 300,
    "num_heads": 6,
    "ffn_dim": 1024,
    "lambda_phys": 0.1  # Reduced to balance total loss
}

# Classes
classes = {
    "ASTHAMA": 0,
    "BRONCHITIS": 1,
    "COPD": 2,
    "HEALTHY": 3,
    "PENUMONIA": 4
}

# Dataset
class RobustAudioDataset(Dataset):
    def __init__(self, file_list, labels, config):
        self.file_list = file_list
        self.labels = labels
        self.config = config
        self.mel_spectrogram = T.MelSpectrogram(
            sample_rate=config["sample_rate"],
            n_fft=1024,
            hop_length=512,
            n_mels=config["n_mels"]
        )
        self.to_db = T.AmplitudeToDB()
        self.invalid_samples = 0

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        try:
            waveform, sample_rate = torchaudio.load(self.file_list[idx])
            if sample_rate != self.config["sample_rate"]:
                waveform = T.Resample(sample_rate, self.config["sample_rate"])(waveform)
            if waveform.shape[0] > 1:
                waveform = waveform.mean(dim=0, keepdim=True)
            mel_spec = self.mel_spectrogram(waveform)
            mel_spec = self.to_db(mel_spec)
            mel_spec = (mel_spec - mel_spec.mean()) / (mel_spec.std().clamp(min=1e-6))
            if mel_spec.shape[-1] > self.config["fixed_length"]:
                mel_spec = mel_spec[:, :, :self.config["fixed_length"]]
            else:
                mel_spec = F.pad(mel_spec, (0, self.config["fixed_length"] - mel_spec.shape[-1]))
            label = self.labels[idx]
            return mel_spec, label
        except Exception as e:
            print(f"Error loading {self.file_list[idx]}: {e}")
            self.invalid_samples += 1
            return torch.zeros(1, self.config["n_mels"], self.config["fixed_length"]), -1

# Model
class EnhancedTransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ffn_dim, dropout):
        super().__init__()
        self.norm1 = nn.LayerNorm(embed_dim)
        self.attention = nn.MultiheadAttention(embed_dim, num_heads, dropout=dropout)
        self.norm2 = nn.LayerNorm(embed_dim)
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, ffn_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(ffn_dim, embed_dim)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        attn_input = self.norm1(x)
        attn_output, _ = self.attention(attn_input, attn_input, attn_input)
        x = x + self.dropout(attn_output)
        ffn_input = self.norm2(x)
        ffn_output = self.ffn(ffn_input)
        x = x + self.dropout(ffn_output)
        return x

class PositionalEncoding(nn.Module):
    def __init__(self, embed_dim, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-np.log(10000.0) / embed_dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:, :x.size(1)]
        return x

class RobustPINNAudioModel(nn.Module):
    def __init__(self, num_classes, config):
        super().__init__()
        embed_dim = config["embed_dim"]
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(32, embed_dim, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(embed_dim),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"])
        )
        self.pos_encoder = PositionalEncoding(embed_dim)
        self.transformer_layers = nn.ModuleList([
            EnhancedTransformerBlock(embed_dim, config["num_heads"], config["ffn_dim"], config["dropout"])
            for _ in range(config["num_layers"])
        ])
        self.classifier = nn.Linear(embed_dim, num_classes)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.conv_layers(x)
        batch_size, c, h, w = x.shape
        x = x.permute(0, 2, 3, 1).reshape(batch_size, h * w, c)
        x = self.pos_encoder(x)
        for layer in self.transformer_layers:
            x = layer(x)
        x = x.mean(dim=1)
        x = self.classifier(x)
        return x

# Loss Function
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, alpha=None, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction

    def forward(self, inputs, targets):
        logpt = F.log_softmax(inputs, dim=-1)
        logpt = logpt.gather(1, targets.view(-1, 1)).view(-1)
        pt = logpt.exp()
        if self.alpha is not None:
            at = self.alpha[targets]
            logpt = logpt * at
        loss = -((1 - pt) ** self.gamma) * logpt
        return loss.mean() if self.reduction == 'mean' else loss.sum()

class PhysicsInformedLoss(nn.Module):
    def __init__(self, alpha=None, lambda_phys=0.1):
        super().__init__()
        self.classification_loss = FocalLoss(gamma=2.0, alpha=alpha)
        self.lambda_phys = lambda_phys

    def spectral_consistency_loss(self, mel_spec):
        freq_diff = torch.diff(mel_spec, dim=2)
        smoothness_loss = torch.mean(freq_diff ** 2)
        energy = torch.mean(mel_spec ** 2, dim=(1, 2))
        energy_loss = torch.var(energy)
        time_diff = torch.diff(mel_spec, dim=3)
        continuity_loss = torch.mean(time_diff ** 2)
        return 0.5 * smoothness_loss + 0.2 * energy_loss + 0.3 * continuity_loss

    def forward(self, outputs, targets, mel_spec):
        class_loss = self.classification_loss(outputs, targets)
        phys_loss = self.spectral_consistency_loss(mel_spec)
        total_loss = class_loss + self.lambda_phys * phys_loss
        return total_loss, class_loss, phys_loss

# Main Setup
torch.manual_seed(42)
np.random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data Loading
def load_and_split_data(data_dir):
    all_files, all_labels = [], []
    for class_name, label in classes.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            warnings.warn(f"Directory {class_dir} does not exist")
            continue
        files = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.endswith('.wav')]
        if not files:
            warnings.warn(f"No .wav files found in {class_dir}")
            continue
        all_files.extend(files)
        all_labels.extend([label] * len(files))
    
    if not all_files:
        raise ValueError("No valid audio files found in the dataset directory")
    
    train_files, temp_files, train_labels, temp_labels = train_test_split(
        all_files, all_labels, test_size=0.3, stratify=all_labels, random_state=42
    )
    val_files, test_files, val_labels, test_labels = train_test_split(
        temp_files, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42
    )
    
    print(f"Train files: {len(train_files)}")
    print(f"Validation files: {len(val_files)}")
    print(f"Test files: {len(test_files)}")
    
    return train_files, val_files, test_files, train_labels, val_labels, test_labels

try:
    train_files, val_files, test_files, train_labels, val_labels, test_labels = load_and_split_data(config["data_dir"])
except Exception as e:
    print(f"Error loading data: {e}")
    exit(1)

class_counts = np.bincount(train_labels + val_labels + test_labels)
print(f"Class counts: {class_counts}")
class_weights = torch.tensor(1.0 / (class_counts + 1e-6), dtype=torch.float).to(device)
class_weights = class_weights / class_weights.sum()

train_dataset = RobustAudioDataset(train_files, train_labels, config)
val_dataset = RobustAudioDataset(val_files, val_labels, config)
test_dataset = RobustAudioDataset(test_files, test_labels, config)

# Training and Evaluation Functions
def evaluate(model, data_loader, criterion, device, return_predictions=False, compute_class_acc=False):
    model.eval()
    total_loss, total_class_loss, total_phys_loss, correct, total = 0.0, 0.0, 0.0, 0, 0
    all_preds, all_targets = [], []
    class_correct = torch.zeros(len(classes)).to(device)
    class_total = torch.zeros(len(classes)).to(device)
    
    with torch.no_grad():
        for inputs, labels in tqdm(data_loader, desc="Evaluating"):
            valid_mask = labels != -1
            if not valid_mask.any():
                continue
            inputs = inputs[valid_mask].to(device)
            labels = labels[valid_mask].to(device)
            outputs = model(inputs)
            loss, class_loss, phys_loss = criterion(outputs, labels, inputs)
            total_loss += loss.item() * inputs.size(0)
            total_class_loss += class_loss.item() * inputs.size(0)
            total_phys_loss += phys_loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            if return_predictions:
                all_preds.extend(predicted.cpu().numpy())
                all_targets.extend(labels.cpu().numpy())
            if compute_class_acc:
                for c in range(len(classes)):
                    class_mask = labels == c
                    class_correct[c] += (predicted[class_mask] == labels[class_mask]).sum().item()
                    class_total[c] += class_mask.sum().item()
    
    avg_loss = total_loss / total if total > 0 else 0
    avg_class_loss = total_class_loss / total if total > 0 else 0
    avg_phys_loss = total_phys_loss / total if total > 0 else 0
    accuracy = correct / total if total > 0 else 0
    class_acc = [class_correct[i] / class_total[i] if class_total[i] > 0 else 0 for i in range(len(classes))]
    
    if return_predictions:
        return avg_loss, avg_class_loss, avg_phys_loss, accuracy, all_preds, all_targets, class_acc
    return avg_loss, avg_class_loss, avg_phys_loss, accuracy

def smooth_curve(points, factor=0.9):
    smoothed = []
    for point in points:
        if smoothed:
            smoothed.append(smoothed[-1] * factor + point * (1 - factor))
        else:
            smoothed.append(point)
    return smoothed

def train_model(model, train_loader, val_loader, test_loader, criterion, optimizer, scheduler, device, config):
    scaler = GradScaler('cuda')
    best_val_acc = 0
    patience_counter = 0
    history = {
        'train_loss': [], 'val_loss': [], 'test_loss': [],
        'train_class_loss': [], 'val_class_loss': [], 'test_class_loss': [],
        'train_phys_loss': [], 'val_phys_loss': [], 'test_phys_loss': [],
        'train_acc': [], 'val_acc': [], 'test_acc': []
    }
    
    for epoch in range(config["epochs"]):
        model.train()
        train_loss, train_class_loss, train_phys_loss, train_correct, train_total = 0.0, 0.0, 0.0, 0, 0
        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
            valid_mask = labels != -1
            if not valid_mask.any():
                continue
            inputs = inputs[valid_mask].to(device)
            labels = labels[valid_mask].to(device)
            optimizer.zero_grad()
            with autocast('cuda'):
                outputs = model(inputs)
                loss, class_loss, phys_loss = criterion(outputs, labels, inputs)
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            scaler.step(optimizer)
            scaler.update()
            train_loss += loss.item() * inputs.size(0)
            train_class_loss += class_loss.item() * inputs.size(0)
            train_phys_loss += phys_loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()
        
        train_loss = train_loss / train_total if train_total > 0 else 0
        train_class_loss = train_class_loss / train_total if train_total > 0 else 0
        train_phys_loss = train_phys_loss / train_total if train_total > 0 else 0
        train_acc = train_correct / train_total if train_total > 0 else 0
        
        val_loss, val_class_loss, val_phys_loss, val_acc, val_preds, val_targets, val_class_acc = evaluate(
            model, val_loader, criterion, device, return_predictions=True, compute_class_acc=True
        )
        test_loss, test_class_loss, test_phys_loss, test_acc = evaluate(model, test_loader, criterion, device)
        
        if epoch < config["warmup_epochs"]:
            lr = config["learning_rate"] * (epoch + 1) / config["warmup_epochs"]
            for param_group in optimizer.param_groups:
                param_group['lr'] = lr
        else:
            scheduler.step()
        
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['test_loss'].append(test_loss)
        history['train_class_loss'].append(train_class_loss)
        history['val_class_loss'].append(val_class_loss)
        history['test_class_loss'].append(test_class_loss)
        history['train_phys_loss'].append(train_phys_loss)
        history['val_phys_loss'].append(val_phys_loss)
        history['test_phys_loss'].append(test_phys_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        history['test_acc'].append(test_acc)
        
        print(f"Epoch {epoch+1}/{config['epochs']} - {train_loader.dataset.invalid_samples} invalid samples")
        print(f"Train Loss: {train_loss:.4f} (Class: {train_class_loss:.4f}, Phys: {train_phys_loss:.4f}) | Acc: {train_acc:.4f}")
        print(f"Val Loss: {val_loss:.4f} (Class: {val_class_loss:.4f}, Phys: {val_phys_loss:.4f}) | Acc: {val_acc:.4f}")
        print(f"Test Loss: {test_loss:.4f} (Class: {test_class_loss:.4f}, Phys: {test_phys_loss:.4f}) | Acc: {test_acc:.4f}")
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save(model.state_dict(), "best_robust_pinn_model.pth")
            print("New best model saved!")
        else:
            patience_counter += 1
            if patience_counter >= config["patience"]:
                print(f"Early stopping at epoch {epoch+1}")
                break
    
    # Plotting
    plt.figure(figsize=(15, 10))
    plt.subplot(2, 2, 1)
    plt.plot(smooth_curve(history['train_loss']), label='Train Loss')
    plt.plot(smooth_curve(history['val_loss']), label='Validation Loss')
    plt.plot(smooth_curve(history['test_loss']), label='Test Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Total Loss')
    plt.legend()
    
    plt.subplot(2, 2, 2)
    plt.plot(smooth_curve(history['train_class_loss']), label='Train Class Loss')
    plt.plot(smooth_curve(history['val_class_loss']), label='Validation Class Loss')
    plt.plot(smooth_curve(history['test_class_loss']), label='Test Class Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Classification Loss')
    plt.legend()
    
    plt.subplot(2, 2, 3)
    plt.plot(smooth_curve(history['train_phys_loss']), label='Train Phys Loss')
    plt.plot(smooth_curve(history['val_phys_loss']), label='Validation Phys Loss')
    plt.plot(smooth_curve(history['test_phys_loss']), label='Test Phys Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Physics Loss')
    plt.legend()
    
    plt.subplot(2, 2, 4)
    plt.plot(smooth_curve(history['train_acc']), label='Train Accuracy')
    plt.plot(smooth_curve(history['val_acc']), label='Validation Accuracy')
    plt.plot(smooth_curve(history['test_acc']), label='Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return history

# Main
def main():
    train_weights = [1.0 / class_counts[label] for label in train_labels]
    sampler = torch.utils.data.WeightedRandomSampler(train_weights, len(train_weights), replacement=True)
    
    train_loader = DataLoader(
        train_dataset, batch_size=config["batch_size"], sampler=sampler,
        num_workers=config["num_workers"], pin_memory=True
    )
    val_loader = DataLoader(
        val_dataset, batch_size=config["batch_size"], shuffle=False,
        num_workers=config["num_workers"], pin_memory=True
    )
    test_loader = DataLoader(
        test_dataset, batch_size=config["batch_size"], shuffle=False,
        num_workers=config["num_workers"], pin_memory=True
    )
    
    model = RobustPINNAudioModel(num_classes=len(classes), config=config).to(device)
    
    criterion = PhysicsInformedLoss(alpha=class_weights, lambda_phys=config["lambda_phys"])
    optimizer = torch.optim.AdamW(model.parameters(), lr=config["learning_rate"], weight_decay=1e-2)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=config["epochs"] - config["warmup_epochs"], eta_min=1e-5)
    
    history = train_model(model, train_loader, val_loader, test_loader, criterion, optimizer, scheduler, device, config)
    
    model.load_state_dict(torch.load("best_robust_pinn_model.pth"))
    train_loss, train_class_loss, train_phys_loss, train_acc, train_preds, train_targets, train_class_acc = evaluate(
        model, train_loader, criterion, device, return_predictions=True, compute_class_acc=True
    )
    test_loss, test_class_loss, test_phys_loss, test_acc, test_preds, test_targets, test_class_acc = evaluate(
        model, test_loader, criterion, device, return_predictions=True, compute_class_acc=True
    )
    
    print("\nFinal Results:")
    print(f"Train Loss: {train_loss:.4f} (Class: {train_class_loss:.4f}, Phys: {train_phys_loss:.4f}) | Acc: {train_acc:.4f}")
    print(f"Test Loss: {test_loss:.4f} (Class: {test_class_loss:.4f}, Phys: {test_phys_loss:.4f}) | Acc: {test_acc:.4f}")
    print("Class-wise Test Accuracy:")
    for class_name, acc in zip(classes.keys(), test_class_acc):
        print(f"  {class_name}: {acc:.4f}")
    
    return history

if __name__ == "__main__":
    history = main()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
from torch.utils.data import DataLoader
from tqdm import tqdm
import torchaudio
import torchaudio.transforms as T
import torch.nn as nn
import torch.nn.functional as F
import os

# Configuration (must match the original training script)
config = {
    "data_dir": "/home/chetak/Speech_Mixing/Cough_Augmented_Dataset",
    "sample_rate": 16000,
    "n_mels": 128,
    "fixed_length": 512,
    "batch_size": 32,
    "num_workers": 4,
    "dropout": 0.3,
    "num_layers": 8,
    "embed_dim": 300,
    "num_heads": 6,
    "ffn_dim": 1024
}

# Classes
classes = {
    "ASTHAMA": 0,
    "BRONCHITIS": 1,
    "COPD": 2,
    "HEALTHY": 3,
    "PENUMONIA": 4
}

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Dataset Class (copied from original script)
class RobustAudioDataset(Dataset):
    def __init__(self, file_list, labels, config):
        self.file_list = file_list
        self.labels = labels
        self.config = config
        self.mel_spectrogram = T.MelSpectrogram(
            sample_rate=config["sample_rate"],
            n_fft=1024,
            hop_length=512,
            n_mels=config["n_mels"]
        )
        self.to_db = T.AmplitudeToDB()
        self.invalid_samples = 0

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        try:
            waveform, sample_rate = torchaudio.load(self.file_list[idx])
            if sample_rate != self.config["sample_rate"]:
                waveform = T.Resample(sample_rate, self.config["sample_rate"])(waveform)
            if waveform.shape[0] > 1:
                waveform = waveform.mean(dim=0, keepdim=True)
            mel_spec = self.mel_spectrogram(waveform)
            mel_spec = self.to_db(mel_spec)
            mel_spec = (mel_spec - mel_spec.mean()) / (mel_spec.std().clamp(min=1e-6))
            if mel_spec.shape[-1] > self.config["fixed_length"]:
                mel_spec = mel_spec[:, :, :self.config["fixed_length"]]
            else:
                mel_spec = F.pad(mel_spec, (0, self.config["fixed_length"] - mel_spec.shape[-1]))
            label = self.labels[idx]
            return mel_spec, label
        except Exception as e:
            print(f"Error loading {self.file_list[idx]}: {e}")
            self.invalid_samples += 1
            return torch.zeros(1, self.config["n_mels"], self.config["fixed_length"]), -1

# Model Classes (copied from original script)
class EnhancedTransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ffn_dim, dropout):
        super().__init__()
        self.norm1 = nn.LayerNorm(embed_dim)
        self.attention = nn.MultiheadAttention(embed_dim, num_heads, dropout=dropout)
        self.norm2 = nn.LayerNorm(embed_dim)
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, ffn_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(ffn_dim, embed_dim)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        attn_input = self.norm1(x)
        attn_output, _ = self.attention(attn_input, attn_input, attn_input)
        x = x + self.dropout(attn_output)
        ffn_input = self.norm2(x)
        ffn_output = self.ffn(ffn_input)
        x = x + self.dropout(ffn_output)
        return x

class PositionalEncoding(nn.Module):
    def __init__(self, embed_dim, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-np.log(10000.0) / embed_dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:, :x.size(1)]
        return x

class RobustPINNAudioModel(nn.Module):
    def __init__(self, num_classes, config):
        super().__init__()
        embed_dim = config["embed_dim"]
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(32, embed_dim, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(embed_dim),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"])
        )
        self.pos_encoder = PositionalEncoding(embed_dim)
        self.transformer_layers = nn.ModuleList([
            EnhancedTransformerBlock(embed_dim, config["num_heads"], config["ffn_dim"], config["dropout"])
            for _ in range(config["num_layers"])
        ])
        self.classifier = nn.Linear(embed_dim, num_classes)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.conv_layers(x)
        batch_size, c, h, w = x.shape
        x = x.permute(0, 2, 3, 1).reshape(batch_size, h * w, c)
        x = self.pos_encoder(x)
        for layer in self.transformer_layers:
            x = layer(x)
        x = x.mean(dim=1)
        x = self.classifier(x)
        return x

# Function to load test data (adapted from original script)
def load_test_data(data_dir):
    all_files, all_labels = [], []
    for class_name, label in classes.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Warning: Directory {class_dir} does not exist")
            continue
        files = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.endswith('.wav')]
        if not files:
            print(f"Warning: No .wav files found in {class_dir}")
            continue
        all_files.extend(files)
        all_labels.extend([label] * len(files))
    
    if not all_files:
        raise ValueError("No valid audio files found in the dataset directory")
    
    # For this script, we only need the test split
    # Assuming the same split as in the original script
    from sklearn.model_selection import train_test_split
    _, temp_files, _, temp_labels = train_test_split(
        all_files, all_labels, test_size=0.3, stratify=all_labels, random_state=42
    )
    _, test_files, _, test_labels = train_test_split(
        temp_files, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42
    )
    
    print(f"Test files: {len(test_files)}")
    return test_files, test_labels

# Function to get predictions and labels
def get_predictions_and_labels(model, data_loader, device):
    model.eval()
    all_probs = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(data_loader, desc="Computing predictions"):
            valid_mask = labels != -1
            if not valid_mask.any():
                continue
            inputs = inputs[valid_mask].to(device)
            labels = labels[valid_mask].to(device)
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)
            all_probs.append(probs.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
    
    all_probs = np.concatenate(all_probs, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    return all_probs, all_labels

# Function to plot ROC curve
def plot_roc_curve(y_true, y_score, classes, title="ROC Curve"):
    # Binarize the labels for multi-class ROC
    y_true_bin = label_binarize(y_true, classes=list(range(len(classes))))
    n_classes = y_true_bin.shape[1]
    
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Compute macro-average ROC curve and ROC area
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= n_classes
    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    
    # Plot all ROC curves
    plt.figure(figsize=(10, 8))
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red'])
    for i, color in zip(range(n_classes), colors):
        plt.plot(
            fpr[i], tpr[i], color=color, lw=2,
            label=f'ROC curve of {list(classes.keys())[i]} (AUC = {roc_auc[i]:.2f})'
        )
    
    plt.plot(
        fpr["macro"], tpr["macro"], color='deeppink', linestyle='--', lw=2,
        label=f'Macro-average ROC curve (AUC = {roc_auc["macro"]:.2f})'
    )
    
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

def main():
    # Load test data
    test_files, test_labels = load_test_data(config["data_dir"])
    
    # Create test dataset and loader
    test_dataset = RobustAudioDataset(test_files, test_labels, config)
    test_loader = DataLoader(
        test_dataset, batch_size=config["batch_size"], shuffle=False,
        num_workers=config["num_workers"], pin_memory=True
    )
    
    # Load model
    model = RobustPINNAudioModel(num_classes=len(classes), config=config).to(device)
    model.load_state_dict(torch.load("best_robust_pinn_model.pth", map_location=device))
    model.eval()
    
    # Get predictions and true labels
    test_probs, test_labels = get_predictions_and_labels(model, test_loader, device)
    
    # Plot ROC curve
    plot_roc_curve(test_labels, test_probs, classes, title="ROC Curve for Multi-Class Classification")

if __name__ == "__main__":
    main()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
from sklearn.preprocessing import label_binarize
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import torchaudio
import torchaudio.transforms as T
import torch.nn as nn
import torch.nn.functional as F
import os
from itertools import cycle

# Configuration (must match the original training script)
config = {
    "data_dir": "/home/chetak/Speech_Mixing/Cough_Augmented_Dataset",
    "sample_rate": 16000,
    "n_mels": 128,
    "fixed_length": 512,
    "batch_size": 32,
    "num_workers": 4,
    "dropout": 0.3,
    "num_layers": 8,
    "embed_dim": 300,
    "num_heads": 6,
    "ffn_dim": 1024
}

# Classes
classes = {
    "ASTHAMA": 0,
    "BRONCHITIS": 1,
    "COPD": 2,
    "HEALTHY": 3,
    "PENUMONIA": 4
}

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Dataset Class (copied from original script)
class RobustAudioDataset(Dataset):
    def __init__(self, file_list, labels, config):
        self.file_list = file_list
        self.labels = labels
        self.config = config
        self.mel_spectrogram = T.MelSpectrogram(
            sample_rate=config["sample_rate"],
            n_fft=1024,
            hop_length=512,
            n_mels=config["n_mels"]
        )
        self.to_db = T.AmplitudeToDB()
        self.invalid_samples = 0

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        try:
            waveform, sample_rate = torchaudio.load(self.file_list[idx])
            if sample_rate != self.config["sample_rate"]:
                waveform = T.Resample(sample_rate, self.config["sample_rate"])(waveform)
            if waveform.shape[0] > 1:
                waveform = waveform.mean(dim=0, keepdim=True)
            mel_spec = self.mel_spectrogram(waveform)
            mel_spec = self.to_db(mel_spec)
            mel_spec = (mel_spec - mel_spec.mean()) / (mel_spec.std().clamp(min=1e-6))
            if mel_spec.shape[-1] > self.config["fixed_length"]:
                mel_spec = mel_spec[:, :, :self.config["fixed_length"]]
            else:
                mel_spec = F.pad(mel_spec, (0, self.config["fixed_length"] - mel_spec.shape[-1]))
            label = self.labels[idx]
            return mel_spec, label
        except Exception as e:
            print(f"Error loading {self.file_list[idx]}: {e}")
            self.invalid_samples += 1
            return torch.zeros(1, self.config["n_mels"], self.config["fixed_length"]), -1

# Model Classes (copied from original script)
class EnhancedTransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ffn_dim, dropout):
        super().__init__()
        self.norm1 = nn.LayerNorm(embed_dim)
        self.attention = nn.MultiheadAttention(embed_dim, num_heads, dropout=dropout)
        self.norm2 = nn.LayerNorm(embed_dim)
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, ffn_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(ffn_dim, embed_dim)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        attn_input = self.norm1(x)
        attn_output, _ = self.attention(attn_input, attn_input, attn_input)
        x = x + self.dropout(attn_output)
        ffn_input = self.norm2(x)
        ffn_output = self.ffn(ffn_input)
        x = x + self.dropout(ffn_output)
        return x

class PositionalEncoding(nn.Module):
    def __init__(self, embed_dim, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-np.log(10000.0) / embed_dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:, :x.size(1)]
        return x

class RobustPINNAudioModel(nn.Module):
    def __init__(self, num_classes, config):
        super().__init__()
        embed_dim = config["embed_dim"]
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(32, embed_dim, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(embed_dim),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"])
        )
        self.pos_encoder = PositionalEncoding(embed_dim)
        self.transformer_layers = nn.ModuleList([
            EnhancedTransformerBlock(embed_dim, config["num_heads"], config["ffn_dim"], config["dropout"])
            for _ in range(config["num_layers"])
        ])
        self.classifier = nn.Linear(embed_dim, num_classes)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.conv_layers(x)
        batch_size, c, h, w = x.shape
        x = x.permute(0, 2, 3, 1).reshape(batch_size, h * w, c)
        x = self.pos_encoder(x)
        for layer in self.transformer_layers:
            x = layer(x)
        x = x.mean(dim=1)
        x = self.classifier(x)
        return x

# Function to load test data (adapted from original script)
def load_test_data(data_dir):
    all_files, all_labels = [], []
    for class_name, label in classes.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Warning: Directory {class_dir} does not exist")
            continue
        files = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.endswith('.wav')]
        if not files:
            print(f"Warning: No .wav files found in {class_dir}")
            continue
        all_files.extend(files)
        all_labels.extend([label] * len(files))
    
    if not all_files:
        raise ValueError("No valid audio files found in the dataset directory")
    
    # For this script, we only need the test split
    from sklearn.model_selection import train_test_split
    _, temp_files, _, temp_labels = train_test_split(
        all_files, all_labels, test_size=0.3, stratify=all_labels, random_state=42
    )
    _, test_files, _, test_labels = train_test_split(
        temp_files, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42
    )
    
    print(f"Test files: {len(test_files)}")
    return test_files, test_labels

# Function to get predictions and labels
def get_predictions_and_labels(model, data_loader, device):
    model.eval()
    all_probs = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(data_loader, desc="Computing predictions"):
            valid_mask = labels != -1
            if not valid_mask.any():
                continue
            inputs = inputs[valid_mask].to(device)
            labels = labels[valid_mask].to(device)
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)
            all_probs.append(probs.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
    
    all_probs = np.concatenate(all_probs, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    return all_probs, all_labels

# Function to plot calibration curves
def plot_calibration_curves(y_true, y_prob, classes, title="Calibration Curves"):
    # Binarize the labels for multi-class calibration
    y_true_bin = label_binarize(y_true, classes=list(range(len(classes))))
    n_classes = y_true_bin.shape[1]
    
    # Compute calibration curves for each class
    prob_true = []
    prob_pred = []
    for i in range(n_classes):
        pt, pp = calibration_curve(y_true_bin[:, i], y_prob[:, i], n_bins=10, strategy='uniform')
        prob_true.append(pt)
        prob_pred.append(pp)
    
    # Compute macro-average calibration curve
    all_prob_true = np.concatenate(prob_true)
    all_prob_pred = np.concatenate(prob_pred)
    macro_prob_true, macro_prob_pred = calibration_curve(
        all_prob_true > 0, all_prob_pred, n_bins=10, strategy='uniform'
    )
    
    # Plot calibration curves
    plt.figure(figsize=(10, 8))
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red'])
    
    # Plot for each class
    for i, color in zip(range(n_classes), colors):
        plt.plot(
            prob_pred[i], prob_true[i], marker='o', color=color, lw=2,
            label=f'{list(classes.keys())[i]}'
        )
    
    # Plot macro-average
    plt.plot(
        macro_prob_pred, macro_prob_true, marker='o', color='deeppink', linestyle='--', lw=2,
        label='Macro-average'
    )
    
    # Plot perfectly calibrated line
    plt.plot([0, 1], [0, 1], color='black', linestyle='--', lw=2, label='Perfectly calibrated')
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('Mean Predicted Probability')
    plt.ylabel('Fraction of Positives')
    plt.title(title)
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

def main():
    # Load test data
    test_files, test_labels = load_test_data(config["data_dir"])
    
    # Create test dataset and loader
    test_dataset = RobustAudioDataset(test_files, test_labels, config)
    test_loader = DataLoader(
        test_dataset, batch_size=config["batch_size"], shuffle=False,
        num_workers=config["num_workers"], pin_memory=True
    )
    
    # Load model
    model = RobustPINNAudioModel(num_classes=len(classes), config=config).to(device)
    model.load_state_dict(torch.load("best_robust_pinn_model.pth", map_location=device))
    model.eval()
    
    # Get predictions and true labels
    test_probs, test_labels = get_predictions_and_labels(model, test_loader, device)
    
    # Plot calibration curves
    plot_calibration_curves(test_labels, test_probs, classes, title="Calibration Curves for Multi-Class Classification")

if __name__ == "__main__":
    main()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import umap
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import torchaudio
import torchaudio.transforms as T
import torch.nn as nn
import torch.nn.functional as F
import os
from itertools import cycle

# Configuration (must match the original training script)
config = {
    "data_dir": "/home/chetak/Speech_Mixing/Cough_Augmented_Dataset",
    "sample_rate": 16000,
    "n_mels": 128,
    "fixed_length": 512,
    "batch_size": 32,
    "num_workers": 4,
    "dropout": 0.3,
    "num_layers": 8,
    "embed_dim": 300,
    "num_heads": 6,
    "ffn_dim": 1024
}

# Classes
classes = {
    "ASTHAMA": 0,
    "BRONCHITIS": 1,
    "COPD": 2,
    "HEALTHY": 3,
    "PENUMONIA": 4
}

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Dataset Class (copied from original script)
class RobustAudioDataset(Dataset):
    def __init__(self, file_list, labels, config):
        self.file_list = file_list
        self.labels = labels
        self.config = config
        self.mel_spectrogram = T.MelSpectrogram(
            sample_rate=config["sample_rate"],
            n_fft=1024,
            hop_length=512,
            n_mels=config["n_mels"]
        )
        self.to_db = T.AmplitudeToDB()
        self.invalid_samples = 0

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        try:
            waveform, sample_rate = torchaudio.load(self.file_list[idx])
            if sample_rate != self.config["sample_rate"]:
                waveform = T.Resample(sample_rate, self.config["sample_rate"])(waveform)
            if waveform.shape[0] > 1:
                waveform = waveform.mean(dim=0, keepdim=True)
            mel_spec = self.mel_spectrogram(waveform)
            mel_spec = self.to_db(mel_spec)
            mel_spec = (mel_spec - mel_spec.mean()) / (mel_spec.std().clamp(min=1e-6))
            if mel_spec.shape[-1] > self.config["fixed_length"]:
                mel_spec = mel_spec[:, :, :self.config["fixed_length"]]
            else:
                mel_spec = F.pad(mel_spec, (0, self.config["fixed_length"] - mel_spec.shape[-1]))
            label = self.labels[idx]
            return mel_spec, label
        except Exception as e:
            print(f"Error loading {self.file_list[idx]}: {e}")
            self.invalid_samples += 1
            return torch.zeros(1, self.config["n_mels"], self.config["fixed_length"]), -1

# Model Classes (copied from original script)
class EnhancedTransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ffn_dim, dropout):
        super().__init__()
        self.norm1 = nn.LayerNorm(embed_dim)
        self.attention = nn.MultiheadAttention(embed_dim, num_heads, dropout=dropout)
        self.norm2 = nn.LayerNorm(embed_dim)
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, ffn_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(ffn_dim, embed_dim)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        attn_input = self.norm1(x)
        attn_output, _ = self.attention(attn_input, attn_input, attn_input)
        x = x + self.dropout(attn_output)
        ffn_input = self.norm2(x)
        ffn_output = self.ffn(ffn_input)
        x = x + self.dropout(ffn_output)
        return x

class PositionalEncoding(nn.Module):
    def __init__(self, embed_dim, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-np.log(10000.0) / embed_dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:, :x.size(1)]
        return x

class RobustPINNAudioModel(nn.Module):
    def __init__(self, num_classes, config):
        super().__init__()
        embed_dim = config["embed_dim"]
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"]),
            nn.Conv2d(32, embed_dim, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(embed_dim),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Dropout(config["dropout"])
        )
        self.pos_encoder = PositionalEncoding(embed_dim)
        self.transformer_layers = nn.ModuleList([
            EnhancedTransformerBlock(embed_dim, config["num_heads"], config["ffn_dim"], config["dropout"])
            for _ in range(config["num_layers"])
        ])
        self.classifier = nn.Linear(embed_dim, num_classes)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x, return_features=False):
        x = self.conv_layers(x)
        batch_size, c, h, w = x.shape
        x = x.permute(0, 2, 3, 1).reshape(batch_size, h * w, c)
        x = self.pos_encoder(x)
        for layer in self.transformer_layers:
            x = layer(x)
        features = x.mean(dim=1)
        if return_features:
            return features
        x = self.classifier(features)
        return x

# Function to load test data (adapted from original script)
def load_test_data(data_dir):
    all_files, all_labels = [], []
    for class_name, label in classes.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Warning: Directory {class_dir} does not exist")
            continue
        files = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.endswith('.wav')]
        if not files:
            print(f"Warning: No .wav files found in {class_dir}")
            continue
        all_files.extend(files)
        all_labels.extend([label] * len(files))
    
    if not all_files:
        raise ValueError("No valid audio files found in the dataset directory")
    
    # For this script, we only need the test split
    from sklearn.model_selection import train_test_split
    _, temp_files, _, temp_labels = train_test_split(
        all_files, all_labels, test_size=0.3, stratify=all_labels, random_state=42
    )
    _, test_files, _, test_labels = train_test_split(
        temp_files, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42
    )
    
    print(f"Test files: {len(test_files)}")
    return test_files, test_labels

# Function to extract features and labels
def get_features_and_labels(model, data_loader, device):
    model.eval()
    all_features = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(data_loader, desc="Extracting features"):
            valid_mask = labels != -1
            if not valid_mask.any():
                continue
            inputs = inputs[valid_mask].to(device)
            labels = labels[valid_mask].to(device)
            features = model(inputs, return_features=True)
            all_features.append(features.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
    
    all_features = np.concatenate(all_features, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    return all_features, all_labels

# Function to plot UMAP visualization
def plot_umap_visualization(features, labels, classes, title="UMAP Visualization of Features"):
    # Apply UMAP
    reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.1)
    umap_results = reducer.fit_transform(features)
    
    # Plot UMAP results
    plt.figure(figsize=(10, 8))
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red'])
    class_names = list(classes.keys())
    
    for class_idx, color in zip(range(len(classes)), colors):
        mask = labels == class_idx
        plt.scatter(
            umap_results[mask, 0], umap_results[mask, 1],
            c=color, label=class_names[class_idx], alpha=0.6, s=50
        )
    
    plt.xlabel('UMAP Component 1')
    plt.ylabel('UMAP Component 2')
    plt.title(title)
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

def main():
    # Load test data
    test_files, test_labels = load_test_data(config["data_dir"])
    
    # Create test dataset and loader
    test_dataset = RobustAudioDataset(test_files, test_labels, config)
    test_loader = DataLoader(
        test_dataset, batch_size=config["batch_size"], shuffle=False,
        num_workers=config["num_workers"], pin_memory=True
    )
    
    # Load model
    model = RobustPINNAudioModel(num_classes=len(classes), config=config).to(device)
    model.load_state_dict(torch.load("best_robust_pinn_model.pth", map_location=device))
    model.eval()
    
    # Extract features and labels
    features, labels = get_features_and_labels(model, test_loader, device)
    
    # Plot UMAP visualization
    plot_umap_visualization(features, labels, classes, title="UMAP Visualization of Feature Representations")

if __name__ == "__main__":
    main()
```
